# Find equilibrium

**Setup model:**

```python
t0 = time.time()
model = TwoAssetModelContClass(name='HANK',like_HANK=True)
print(f'model created in {time.time()-t0:.1f} secs')
```

**Solve for monopolistic competition:**

```python
model.find_ra_mon_comp(KN0=28.16527153,Pi0=0.17397939,tol=1e-5) # using good starting values
```

**Results:**

```python
model.show_moments()
```

**Solve for perfect competition:**

```python
model.find_ra_perf_comp(KN0=33.16529328,tol=1e-5)
```

**Results:**

```python
model.show_moments()
```

**Results for same mean labor efficiency as in monopolistic competition:**

```python
model.find_ra_perf_comp(KN0=33.16529328,tol=1e-5,fix_zeta=True)
model.show_moments()
```

# Steady-state comparison for different levels of $B/Y$

**Results for varying zeta:**

```python
sim_data = model.vary_param_GE_perf_comp('rb',[0.002,0.003,0.005,0.007,0.008],KN0=33.225,tol=1e-5)
```

```python
from matplotlib import pyplot as plt

fig1 = plt.figure(figsize=(12,6))
fig2 = plt.figure(figsize=(12,6))
fig3 = plt.figure(figsize=(18,6))
x = sim_data['BY']

# a. Interest rate plots

ax = fig1.add_subplot(2,2,1)
ax.grid()
ax.plot(sim_data['rb'],sim_data['ra'],'bo')
ax.set_xticks(sim_data['rb'])
ax.set_xlabel(r'$r_b$')
ax.set_ylabel(r'$r_a$')
ax.set_title(r'$r_a(r_b)$')

ax = fig1.add_subplot(2,2,2)
ax.grid()
ax.plot(x,sim_data['ra']-sim_data['rb'],'bo')
ax.set_xticks(x)
ax.set_xlabel(r'$B/Y$')
ax.set_ylabel(r'$r_a-r_b$')
ax.set_title('Interest rate spread')

ax = fig1.add_subplot(2,2,3)
ax.grid()
ax.plot(x,sim_data['ra'],'bo')
ax.set_xticks(x)
ax.set_xlabel(r'$B/Y$')
ax.set_ylabel(r'$r_a$')
ax.set_title('Illiquid asset rate')

ax = fig1.add_subplot(2,2,4)
ax.grid()
ax.plot(x,sim_data['rb'],'bo')
ax.set_xticks(x)
ax.set_xlabel(r'$B/Y$')
ax.set_ylabel(r'$r_b$')
ax.set_title('Liquid asset rate')

# b. Asset plots
ax = fig2.add_subplot(2,2,1)
ax.grid()
ax.plot(x,sim_data['KY'],'bo')
ax.set_xticks(x)
ax.set_xlabel(r'$B/Y$')
ax.set_ylabel(r'$K/Y$')
ax.set_title(r'$K/Y(B/Y)$')

ax = fig2.add_subplot(2,2,2)
ax.grid()
ax.plot(x,sim_data['a_gini'],'bo')
ax.set_xticks(x)
ax.set_xlabel(r'$B/Y$')
ax.set_ylabel('Gini illiquid assets')
ax.set_title('Gini illiquid assets')

ax = fig2.add_subplot(2,2,3)
ax.grid()
ax.plot(x,sim_data['b_gini'],'bo')
ax.set_xticks(x)
ax.set_xlabel(r'$B/Y$')
ax.set_ylabel('Gini liquid assets')
ax.set_title('Gini liquid assets')

ax = fig3.add_subplot(2,3,1)
ax.grid()
ax.plot(x,sim_data['c'],'bo')
ax.set_xticks(x)
ax.set_xlabel(r'$B/Y$')
ax.set_ylabel('Consumption')
ax.set_title('Consumption')

ax = fig3.add_subplot(2,3,2)
ax.grid()
ax.plot(x,sim_data['c_gini'],'bo')
ax.set_xticks(x)
ax.set_xlabel(r'$B/Y$')
ax.set_ylabel('Consumption Gini')
ax.set_title('Consumption Gini')

ax = fig3.add_subplot(2,3,3)
ax.grid()
ax.plot(x,sim_data['v'],'bo')
ax.set_xticks(x)
ax.set_xlabel(r'$B/Y$')
ax.set_ylabel('Value function')
ax.set_title('Value function')

ax = fig3.add_subplot(2,3,4)
ax.grid()
ax.plot(x,sim_data['v_planner'],'bo')
ax.set_xticks(x)
ax.set_xlabel(r'$B/Y$')
ax.set_ylabel('Value function, planner')
ax.set_title('Value function, planner')

ax = fig3.add_subplot(2,3,5)
ax.grid()
ax.plot(x,sim_data['MPC'],'bo')
ax.set_xticks(x)
ax.set_xlabel(r'$B/Y$')
ax.set_ylabel('MPC out of $500')
ax.set_title('MPC out of $500')

fig1.tight_layout()
fig2.tight_layout()
fig3.tight_layout()
```
